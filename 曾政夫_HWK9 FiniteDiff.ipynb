# Homework

Consider the following function:

\begin{align}
 f(x) = \log(x)\times \exp\left[-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2 \right].
\end{align}

Let's assume $\mu=1.5$ and $\sigma=2$. 

```julia
μ = 1.5
σ = 2
f(x) = log(x) * exp(-0.5* ((x- μ)/ σ)^ 2)
```

```julia
f_prime(x,h) = (f(x+h) - f(x)) /h
true_f_prime(x) = 1/x *exp(-0.5* ((x- μ)/ σ)^ 2)+log(x) * exp(-0.5* ((x- μ)/ σ)^ 2) * - ((x- μ)/ σ) * 1/σ
```

```julia
true_f_prime(1.2)
```

```julia
ϵ = sqrt(eps(one(eltype(1.0)))) 
h(x) = ϵ *(abs(x)+ ϵ )
```

### Write a numerical finite differentiation program using the forward difference formula. 
#### Calculate the derivative of $f(x)$ at $x=1.2$ using various values of $h$, including Nash's formula. Which one is the best?

```julia
for i in  [1e-4, 1e-5, 1e-6, 1e-7, 1e-8, 1e-9, 1e-10, 1e-11, 1e-12, 1e-13, 1e-14, ϵ, h(1.2)]
    println("h =",i,"\t; derivative of  f(x) at  x=1.2: ",f_prime(1.2,i),"\t deviate : ", abs(f_prime(1.2,i)-true_f_prime(1.2)))
end
```

```julia
function deviate(h,f::Function)
    return abs(f(1.2,h) - true_f_prime(1.2))
end
@manipulate for h_node in (10,100,1000,10000)
    i = range(1.7e-8,1.8e-8,length= h_node)
    plot(i,deviate.(i,f_prime))
end
```

### Write a numerical finite differentiation program using the central difference formula. 
#### Calculate the derivative of $f(x)$ at $x=1.2$ using various values of $h$, including Nash's formula. Which one is the best?

```julia
function centrol_diff(x,h)
    return (f(x + h/2) - f(x - h/2))/h
end
```

```julia
for i in  [1e-4, 1e-5, 1e-6, 1e-7, 1e-8, 1e-9, 1e-10, 1e-11, 1e-12, 1e-13, 1e-14, ϵ, h(1.2)]
    println("h =",i,"\t; derivative of  f(x) at  x=1.2: ",centrol_diff(1.2,i),"\t deviate : ", abs(centrol_diff(1.2,i)-true_f_prime(1.2)))
end
```

h = 𝑒_𝑟𝑜𝑜𝑡×(1.2+𝑒_𝑟𝑜𝑜𝑡) 在幾個代表性的點中是最好的。

```julia
@manipulate for h_node in (10,100,1000,10000)
    i = range(1.7e-8,1.8e-8,length= h_node)
    plot(i,deviate.(i,centrol_diff))
end
```

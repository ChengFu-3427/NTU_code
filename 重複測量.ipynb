## problem3

Random intercept model 與 R‐side analysis 相通 ，其相當於在 R‐side analysis 中將R矩陣指定為對角線是 1 其餘是$\frac{\sigma_{\gamma}^2}{\sigma_{\gamma}^2+\sigma_{\epsilon}^2}$ 的可交換相關係數矩陣。  
因為 R‐side analysis 允許 error term 彼此之間有 correlation ，而 Random intercept model 組內的任兩個觀測值有相同的相關係數，所以 Random intercept model 中的 G‐side analysis 與 R‐side analysis 相通。 

## SAS CODE
   
proc mixed data= mathachieve covtest;  
class School;  
model MathAch= SES / s;  
random intercept/ subject=School G  vcorr=135 s  ;  
run;  
    
proc mixed data= mathachieve covtest;  
class School;  
model MathAch= SES / s;  
random intercept SES/ subject=School G type=un  vcorr=135 s;   
run;

Ex 2: In regression analysis, if you observe an F test statistic = 3.2 (right-tailed test), and under the null hypothesis this statistic should follow an F distribution with a numerator d.f. of 3 and a denominator d.f. of 194, please find the p-value? Please plot the pdf and cdf of this F distribution.

```r
x <-seq(0,6,0.01)
y <-df(x,3,194)
plot(x,y,type="l",main="pdf")
```

```r
x <-seq(0,6,0.01)
y <-pf(x,3,194)
plot(x,y,type="l",main="cdf")
```

```r
ex2p_value <-1-pf(3.2,3,194)
ex2p_value
```

p-value = 0.0244858295814092

Ex 3: In regression analysis, if you observe a t test statistic = -2.08 (two-tailed test), and under the null hypothesis this statistic should follow a t distribution with a d.f. of 136, please find the p-value?

```r
p_value <- pt(-2.08,136)*2
p_value
```

p_value = 0.039402674094279 (t test statistic = -2.08)

What’s the p-value if you observe a t test statistic = 2.45 (two-tailed test). Please plot the pdf and cdf of this t distribution.

```r
x <-seq(-4,4,0.01)
y <-dt(x,136)
plot(x,y,type="l",main="pdf")
```

```r
x <-seq(-4,4,0.01)
y <-pt(x,136)
plot(x,y,type="l",main="cdf")
```

```r
p_value <- (1-pt(2.45,136))*2
p_value
```

p_value = 0.0155564199625604 (t test statistic = 2.45)

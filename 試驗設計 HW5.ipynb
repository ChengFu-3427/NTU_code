### Problem 1

```r
y <- c(
      33.6,31.1,33.0,28.4,31.4,
      37.1,34.5,29.5,29.9,28.3,
      34.1,30.5,29.2,31.6,28.9,
      34.6,32.7,30.7,32.3,28.6,
      35.4,30.7,30.7,28.1,29.6,
      36.1,30.3,27.9,26.9,33.4
)
block <- rep(c(1,2,3,4,5,6),each=5)
inch <- rep(c(18,24,30,36,42),6)
data <- data.frame(
    inch=inch,
    block=factor(block),
    y=y
)
data
```

```r
data$inch<-as.factor(data$inch)
contrastmatrix<-contr.poly(5)
contrasts(data$inch)<-contrastmatrix
mod<-aov(y~inch+block,data)
summary(mod, split = list(inch= list("linear" = 1, "quadratic" = 2,"cubic"=3,"quartic"=4)))
```

經過以上的分析，我們能夠做出row spacing對soybean的產量在顯著水準為0.05時有線性和二次的關係的結論。

### Problem 2

```r
y <- c(
    10,12,14,
    16,19,21,
    24,27,32,
    12,15,17,
    19,23,29,
    33,34,37,
    24,27,29,
    39,41,43,
    45,47,52
)
treat <- rep(c("a0b0","a0b1","a0b2","a1b0","a1b1","a1b2","a2b0","a2b1","a2b2"),each=3)
a <- rep(c(0,1,2),each=9)
b <- rep(rep(c(0,1,2),each=3),3)
ab <- rep(c(0,0,0,0,1,2,0,3,4),each=3)
rep <- rep(c(1,2,3),9)
data2 <- data.frame(
    a=factor(a),
    b=factor(b),
    ab=factor(ab),
    treat=factor(treat),
    rep=factor(rep),
    y=y
)
data2
```

```r
contrasts(data2$a)<-contr.poly(3)
contrasts(data2$b)<-contr.poly(3)
contrasts(data2$ab)<-contr.poly(5)
mod2<-aov(y~rep+a*b,data2)
summary.lm(mod2)
```

在顯著水準為0.05時，經過以上的分析，我們能夠做出以下結論：  
treatment A對觀測值有線性和二次的關係。  
treatment B對觀測值有線性關係。  
treatment AxB除了a.Q:b.L外，皆和觀測值有關係。

```r
mod3<-aov(y~rep+a*b*c,data3)
summary(mod3)
```

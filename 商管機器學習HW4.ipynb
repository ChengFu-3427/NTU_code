# 1.(House  price, 25%)
In  this  question,  we  use  the  house  price  dataset  of house_sales_data() in the Julia package VMLS. Suppose the house price follows 

```julia
X=house_sales_data()
y=X["price"]
area=X["area"]
areasq=X["area"].*X["area"]
bed=X["beds"]
location1=copy(X["location"])
location2=copy(X["location"])
location3=copy(X["location"])
for i in 1:length(X["location"])
    if (X["location"][i] == 1)
        location1[i]=1
        location2[i]=0
        location3[i]=0
    elseif (X["location"][i] == 2)
        location1[i]=0
        location2[i]=1
        location3[i]=0
    elseif (X["location"][i] == 3)
        location1[i]=0
        location2[i]=0
        location3[i]=1
    else
        location1[i]=0
        location2[i]=0
        location3[i]=0
    end
end
A=hcat(ones(length(y)),area,areasq,bed,location1,location2,location3)
```

# 2.(Cross-validation, 25%) 
Following  question  1,use  5  randomly  chosen  folds  and separate the data into training and testing datasets. (You maychoose a reasonable number of data in each fold.)

```julia
X=house_sales_data()
y=copy(X["price"])
A=hcat(ones(length(y)),area,areasq,bed,X["location"].==1,X["location"].==2,X["location"].==3)
nfold = div(length(y),5)
I = Random.randperm(length(y))
b = zeros(5,7)
error_mat = zeros(5,2)
for i =1:5
    if i in 1:4
        test=I[(nfold*(i-1))+1:nfold*(i)]
        train=setdiff(I,test)
    else
        test=I[(nfold*(i-1))+1:end]
        train=setdiff(I,test)
    end
    coef = A[train,:] \ y[train]
    yhat_train = A[train,:] * coef
    error_mat[i,1]=sqrt(((y[train]-yhat_train)'*(y[train]-yhat_train))/length(yhat_train))
    yhat_test = A[test,:] * coef
    error_mat[i,2]=sqrt(((y[test]-yhat_test)'*(y[test]-yhat_test))/length(yhat_test))
    b[i,:]=coef
end
```

(a)

```julia
Ans = hcat(b,error_mat)
df = DataFrame( beta0 = Ans[:,1], beta1 = Ans[:,2], beta2 = Ans[:,3], beta3 = Ans[:,4], 
    beta4 = Ans[:,5],beta5 = Ans[:,6],beta6 = Ans[:,7] )
```

# 4.(Time series, 25%)
In this question, we use the dataset temperature_data()from the packageVMLS.Build an auto-regressive time series modelwith different memoryvalues.

```julia
function auto_reg(m,day=100)
    T=temperature_data()
    y=T[m+1:end]
    Z=vcat([reverse(T[i-m+1:i]) for i = m:length(T)-1]' ... )
    theta=Z \ y
    predict= Z* theta
    d=day
    rms_error=sqrt(((y-predict)'*(y-predict))/length(y))
    println("RMSE:",rms_error)
    scatter(1:d,T, label =nothing)
    plot!(m+1:d,predict[1:d-m],label="predict",color="black")   
end
```

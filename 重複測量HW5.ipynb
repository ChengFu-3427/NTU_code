```r
setwd("C:/Users/User/Desktop")
data <- read.csv("milk.csv")
```

```r
Y <- matrix(data$milk, ncol=1)
X <- rep(1,length(Y))
Z <- matrix(0,length(Y),10)
for(i in 1:10){
 Z[(12*i-11):(12*i),i] <- 1
}
G <- diag(7.8247,10)
R <- diag(6.6729,length(Y))
```

```r
coef_var <- rbind(
cbind(t(X)%*%solve(R)%*%X,t(X)%*%solve(R)%*%Z),
cbind(t(Z)%*%solve(R)%*%X,(t(Z)%*%solve(R)%*%Z+solve(G)))
)
coef_var <- solve(coef_var)
sqrt(diag(coef_var))[1] #beta標準誤
```

```r
coef[1] #beta估計值
```

```r
sqrt(diag(coef_var))[2:length(sqrt(diag(coef_var)))] #gamma標準誤
```

```r
coef <- coef_var%*%rbind(t(X)%*%solve(R)%*%Y,t(Z)%*%solve(R)%*%Y)
coef[2:length(coef)] #gamma係數估計值
```

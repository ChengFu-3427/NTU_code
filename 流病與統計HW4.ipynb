Ex S1: Please calculate “factorial(10)” by using a “for” loop and a “while” loop, respectively.

```r
for_factorial <- function (x){
    ans <- 1
    for (i in 1:x){
        ans <- ans * i
    }
    return(ans)
}
```

```r
for_factorial(10)
```

```r
while_factorial <- function (x){
    ans <- 1
    done <- FALSE
    i <- 1
    while(done != TRUE){
        ans <- ans*i
        if(i==x)
            done <- TRUE 
        i <- i+1
    }
    return(ans)
}
```

```r
while_factorial(10)
```

```r
x <-c(3600, 5000, 12000, NA, 1000, 2000, 600, 7500, 1800, 9000)
done <- FALSE
i <- 1
while(done!=TRUE){
    if(is.na(x[i])==TRUE){
        done <- TRUE
        cat(i,"\n")        
    }

    i <- i+1
}
```

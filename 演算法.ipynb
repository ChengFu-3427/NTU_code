```julia
using Plots
```

```julia
f(n) = n^0.01
g(n) = log(n)
i(n) = f(n)/g(n)
```

```julia
plot([n  for n in (100:200)],[i(n) for n in (100:200)])
```

```julia
f(n) = n^0.5/log(n)^4
g(n) = n^(5/6)
i(n) = f(n)/g(n)
plot([n  for n in (100:200)],[i(n) for n in (100:200)])
```
